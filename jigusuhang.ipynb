```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('상관용.xlsx')

data
```

```python
data_a = data[['면적(ha)','평균기온(℃)']]
corr1 = data_a.corr(method = 'pearson')
corr1
```

```python
data_b = data[['생산량(톤)','평균기온(℃)']]
corr2 = data_b.corr(method = 'pearson')
corr2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

name1 = ['면적(ha)',	'평균기온(℃)']
sns.set(font_scale=2.0)

plt.rcParams['font.family'] = 'NanumBarunGothic'
sns.set(font="NanumBarunGothic",
rc={"axes.unicode_minus":False}, style='white')
test_heatmap = sns.heatmap(corr1.values, # 데이터
                          cbar = True, # 오른쪽 컬러 막대 출력 여부
                           annot = True, # 차트에 숫자를 보여줄 것인지 여부
                          annot_kws={'size' : 20}, # 숫자 출력 시 숫자 크기 조절
                           fmt = '.4f', # 숫자의 출력 소수점자리 개수 조절
                           square = 'True', # 차트를 정사각형으로 할 것인지
                          yticklabels=name1, # y축에 컬럼명 출력
                          xticklabels=name1) # x축에 컬럼명 출력
plt.title('제주도의 망고 재배면적과 평균 기온의 상관관계', fontsize=15, pad=5)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

name2 = ['생산량(톤)',	'평균기온(℃)']
# 레이블의 폰트 사이즈를 조정
sns.set(font_scale=2.0)

plt.rcParams['font.family'] = 'NanumBarunGothic'
sns.set(font="NanumBarunGothic",
rc={"axes.unicode_minus":False}, style='white')
test_heatmap = sns.heatmap(corr2.values, # 데이터
                          cbar = True, # 오른쪽 컬러 막대 출력 여부
                           annot = True, # 차트에 숫자를 보여줄 것인지 여부
                          annot_kws={'size' : 20}, # 숫자 출력 시 숫자 크기 조절
                           fmt = '.4f', # 숫자의 출력 소수점자리 개수 조절
                           square = 'True', # 차트를 정사각형으로 할 것인지
                          yticklabels=name2, # y축에 컬럼명 출력
                          xticklabels=name2) # x축에 컬럼명 출력
plt.title('제주도의 망고 생산량과 평균 기온의 상관관계', fontsize=15, pad=5)
plt.tight_layout()
plt.show()
```

```python
XT = data[["년도"]]
yT = data["평균기온(℃)"]
```

```python
from sklearn.model_selection import train_test_split

XT_train, XT_test, yT_train, yT_test = train_test_split(XT, yT, test_size=0.2)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

lrT = LinearRegression()
lrT.fit(XT_train, yT_train)
yT_pred = lrT.predict(XT_test)
r2_score(yT_test, yT_pred)
```

```python
plt.scatter(XT_test["년도"], yT_test, label='test')
plt.scatter(XT_test["년도"], yT_pred, c='r', label='predict')
plt.legend()
plt.show()
```

```python
import torch

T_list=[]
for i in range(2023,2051):
  Tpred = lrT.predict([[i]])
  T_list.append(Tpred)

x = torch.linspace(2023,2050,28)
plt.figure(figsize = (8,4))
plt.xlabel('연도')
plt.ylabel('평균기온(℃)')
plt.title('2050년 까지 연도별 평균기온 변화 예측')
plt.grid(True)
plt.plot(x, T_list, marker='o')
```

```python
XM = data[["년도", "평균기온(℃)"]]
yM = data["생산량(톤)"]

from sklearn.model_selection import train_test_split

XM_train, XM_test, yM_train, yM_test = train_test_split(XM, yM, test_size=0.2)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

lrM = LinearRegression()
lrM.fit(XM_train, yM_train)
yM_pred = lrM.predict(XM_test)
r2_score(yM_test, yM_pred)
```

```python
plt.scatter(XM_test["년도"], yM_test, label='test')
plt.scatter(XM_test["년도"], yM_pred, c='r', label='predict')
plt.legend()
plt.show()
```
